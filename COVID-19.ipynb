```python

from statsmodels.tsa.stattools import adfuller

result=adfuller(train.Confirmed)
print('ADF',result[0])
print('pvalue',result[1])
print('No of lag',result[2])
print('The number of observations used for the ADF regression and calculation of the critical values',result[3])
print('Critical value',result[4])
for key,val in result[4].items():
    print("\t",key,": ",val)
```

As per AD Fuller data is non stationary and reject the Null Hypothesis

### Multiple Input series

```python
# training data
def split_sequence(sequence,n_steps):
    X,Y=list(),list()
    for i in range(len(sequence)):
        #find the end of this pattern
        end_ix = i + n_steps
        if end_ix > len(sequence):
            break
        #input and output sample data
        seq_x, seq_y = sequence[i:end_ix,:-1], sequence[end_ix-1,-1]
        print(seq_x,seq_y)
        X.append(seq_x)
        Y.append(seq_y)
    return np.array(X),np.array(Y)
            
    
train["confirmed_diff"]=train.Confirmed.diff(1)
train=train.dropna()
sequence=np.array( train[["recovered","confirmed_diff","Confirmed"]])
X,Y=split_sequence(sequence,1)
n=X.shape[1]*X.shape[2]
X=X.reshape((X.shape[0],n))
```

```python
model.fit(X,Y,epochs=100,verbose=1) # train the model
```
